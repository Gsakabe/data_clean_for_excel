```python
import pandas as pd

import warnings
warnings.filterwarnings('ignore')
```

```python
def refine_df(df):
    col = ["ind","dom","stay","stay_tour","stay_home","stay_work","day","day_tour","day_home","day_work","abr","abr_tour","abr_home","abr_work"]
    ind = ["Hokakido","Touhoku","Kanto","Hokuriku","Tyuubu","Kinki","Tyuugoku","Shikoku","Kyusyu","Okinawa","Asia","Europ","N_America","S_America","Africa","Oceania"]
    df.drop([0,1,2,3,4],inplace=True)
    df.columns = col
    df = df.head(16)
    df.drop(["dom","stay","day","abr"],axis=1,inplace=True)

    df.loc[15,"stay_tour"] = df.loc[15,"abr_tour"]
    df.loc[15,"stay_home"] = df.loc[15,"abr_home"]
    df.loc[15,"stay_work"] = df.loc[15,"abr_work"]

    df.loc[16,"stay_tour"] = df.loc[16,"abr_tour"]
    df.loc[16,"stay_home"] = df.loc[16,"abr_home"]
    df.loc[16,"stay_work"] = df.loc[16,"abr_work"]

    df.loc[17,"stay_tour"] = df.loc[17,"abr_tour"]
    df.loc[17,"stay_home"] = df.loc[17,"abr_home"]
    df.loc[17,"stay_work"] = df.loc[17,"abr_work"]

    df.loc[18,"stay_tour"] = df.loc[18,"abr_tour"]
    df.loc[18,"stay_home"] = df.loc[18,"abr_home"]
    df.loc[18,"stay_work"] = df.loc[18,"abr_work"]

    df.loc[19,"stay_tour"] = df.loc[19,"abr_tour"]
    df.loc[19,"stay_home"] = df.loc[19,"abr_home"]
    df.loc[19,"stay_work"] = df.loc[19,"abr_work"]

    df.loc[20,"stay_tour"] = df.loc[20,"abr_tour"]
    df.loc[20,"stay_home"] = df.loc[20,"abr_home"]
    df.loc[20,"stay_work"] = df.loc[20,"abr_work"]
    
    df.drop(["abr_tour","abr_home","abr_work","ind"],axis=1,inplace=True)
    df.replace({"-" : 0, "â€¦" : 0},inplace=True)
    df.index = ind
    
    return df
```

```python
df_201810 = refine_df(pd.read_excel('201810.xlsx',sheet_name=3)).T
df_201807 = refine_df(pd.read_excel('201807.xlsx',sheet_name=3)).T
df_201804 = refine_df(pd.read_excel('201804.xlsx',sheet_name=3)).T
df_201801 = refine_df(pd.read_excel('201801.xlsx',sheet_name=3)).T
df_201710 = refine_df(pd.read_excel('201710.xls',sheet_name=3)).T
df_201707 = refine_df(pd.read_excel('201707.xls',sheet_name=3)).T
df_201704 = refine_df(pd.read_excel('201704.xls',sheet_name=3)).T
df_201701 = refine_df(pd.read_excel('201701.xls',sheet_name=3)).T
```

```python
def to_Series(df,col):
    se = df.stack().reset_index(name=col)
    ind = se["level_0"] + "_" + se["level_1"]
    se.drop(["level_0","level_1"],axis=1,inplace=True)
    se.index = ind
    
    return se
```

```python
df_201810 = to_Series(df_201810,"201810")
df_201807= to_Series(df_201807,"201807")
df_201804 = to_Series(df_201804,"201804")
df_201801 = to_Series(df_201801,"201801")
df_201710 = to_Series(df_201710,"201710")
df_201707 = to_Series(df_201707,"201707")
df_201704 = to_Series(df_201704,"201704")
df_201701 = to_Series(df_201701,"201701")
```
